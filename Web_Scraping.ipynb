```python
from bs4 import BeautifulSoup
import requests
```

### Reading The Data

```python
url = "https://www.nirfindia.org/2023/EngineeringRanking.html"
page = requests.get(url)
```

```python
soup = BeautifulSoup(page.text,'html')
```

```python
print(soup)
```

#### Finding the data of table 

```python
soup.find('table')
```

```python
soup.find_all("table")
```

```python
table = soup.find_all('table',class_ = 'table table-condensed')
table
```

```python
 table_title = soup.find_all('th')
```

#### Printing the Attributes 

```python
for col in table_title[:15]:
    print(col.text)
```

```python
table_col_name = [col.text.strip() for col in table_title]
```

```python
table_col_name
```

```python
table_col_name = table_col_name[:11]
```

```python
table_col_name
```

```python
table_col_name = ['Institute ID',
 'Name',
 'TLR (100)',
 'RPC (100)',
 'GO (100)',
 'OI (100)',
 'PERCEPTION (100)',
 'City',
 'State',
 'Score',
 'Rank']
```

```python
import pandas as pd
```

### Creating Table using Pandas

```python
df = pd.DataFrame(columns = table_col_name)
```

```python
df
```

```python
table_row_data = soup.find_all('tr')
```

```python
table_row_data
```

### Adding the Rows Data

```python
for row in table_row_data[1:]:
    data_row = row.find_all('td')
    individual_row_data = [i.text.strip() for i in data_row]
    print(individual_row_data)
    if individual_row_data != [] and len(individual_row_data) == len(table_col_name):
        # Create a new row in the DataFrame
        remove_data = individual_row_data[1]
        individual_row_data[1] = remove_data.split('More Details')[0].strip() #Removing unwanted data sentences "More DetailsClose ... .. ."
        df.loc[len(df)] = individual_row_data
    
```
